```r
# Install packages
install.packages("ggbreak")
```

```r
# Load the libraries
library(tidyverse)
library(ggbreak)
```

```r
# Create the dataframe
groups <- c("A", "B", "C", "D")
values <- c(11000, 200, 50, 10)
df <- data.frame(
  groups = groups,
  values = values
)
df
```

```r
# Visualize without axis break
options(repr.plot.width = 12, repr.plot.height = 8)
df |>
  ggplot(aes(x = reorder(groups, values), y = values)) +
    geom_bar(stat = "identity", position = "dodge", fill = "dodgerblue", width = 0.6) +
    geom_text(aes(label = values),
              hjust = -0.05,
              fontface = "bold",
              size = 8) +
    labs(x = "groups",
     y = "values",
     title = "Plot without axis break") +
    coord_flip() +
    scale_y_continuous(limits = c(0, max(values) + 500), breaks = seq(0, max(values), 1000)) +
    theme_bw() +
    theme(axis.text = element_text(size = 22),
          axis.title = element_text(size = 28),
          title = element_text(size = 32))
```

```r
# Visualize with axis break
options(repr.plot.width = 10, repr.plot.height = 8)
df |>
  ggplot(aes(x = reorder(groups, values), y = values)) +
    geom_bar(stat = "identity", position = "dodge", fill = "orangered", width = 0.6) +
    geom_text(aes(label = values),
              hjust = -0.05,
              fontface = "bold",
              size = 8) +
    labs(x = "groups",
     y = "values",
     title = "Plot with axis break") +
    coord_flip() +
    scale_y_continuous(limits = c(0, max(values) + 100), breaks = seq(0, max(values), 100)) +
    scale_y_break(c(250, 10900), scales = 0.3) +
    theme_bw() +
    theme(axis.text = element_text(size = 22),
          axis.title = element_text(size = 28),
          title = element_text(size = 32))
```
